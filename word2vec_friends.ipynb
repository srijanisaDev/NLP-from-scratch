```python
import pandas as pd
import numpy as np
```

```python
import gensim
import os
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []

with open('Friends_Transcript.txt', encoding='latin-1') as f:
    corpus = f.read()

raw_sent = sent_tokenize(corpus)

for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
# story
```

```python
len(story)
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(
    sentences=story,
    vector_size=100,      # 100-dimensional vectors
    window=10,             # 10-word context window
    min_count=2,          # Ignore words appearing less than 2 times
    sg=1,                 # Skip-gram algorithm
    epochs=5            # Train 5 times over data
)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('monica')
```
